```python
import os

def dump_dir_contents(path, indent=""):
    """
    Recursively list directory contents and dump file contents.
    
    Args:
        path: Path to the directory or file
        indent: Indentation string for nested directories
    """
    if os.path.isdir(path):
        # List directory contents
        try:
            items = os.listdir(path)
        except PermissionError:
            print(f"{indent}[Permission denied: {path}]")
            return
            
        for item in sorted(items):
            item_path = os.path.join(path, item)
            
            if os.path.isdir(item_path):
                print(f"{indent}ðŸ“ {item}/")
                dump_dir_contents(item_path, indent + "  ")
            else:
                print(f"{indent}ðŸ“„ {item}")
                dump_file_contents(item_path, indent + "  ")
    else:
        # It's a single file
        print(f"ðŸ“„ {os.path.basename(path)}")
        dump_file_contents(path, "")

def dump_file_contents(file_path, indent):
    """
    Dump file contents with error handling.
    
    Args:
        file_path: Path to the file
        indent: Indentation string
    """
    try:
        # Check if file is binary by trying to read as text
        with open(file_path, 'r', encoding='utf-8') as f:
            # Try reading a small chunk to test encoding
            chunk = f.read(1024)
            f.seek(0)  # Reset to beginning
            
            # Print file content with separator
            print(f"{indent}--- Content of {os.path.basename(file_path)} ---")
            content = f.read()
            if content.strip():  # Only print if there's content
                print(f"{indent}{content}")
            else:
                print(f"{indent}[Empty file]")
            print(f"{indent}{'-' * (40 + len(os.path.basename(file_path)))}")
            
    except UnicodeDecodeError:
        # File is likely binary
        try:
            size = os.path.getsize(file_path)
            print(f"{indent}[Binary file - {size:,} bytes]")
        except:
            print(f"{indent}[Binary file]")
            
    except PermissionError:
        print(f"{indent}[Permission denied]")
        
    except Exception as e:
        print(f"{indent}[Error reading file: {e}]")

dump_dir_contents('.\\gu_toolkit')
  
```
