```python
# Generator of random sums of sin functions
def make_random_sin_function(modes=5):
    expr=0
    for n in range(5):
        # Generate random integer j from 0 to 20
        # Calculate coefficient: j/10 - 1
        j = np.random.randint(0, 20)
        a_n = sp.Rational(j, 10) - 1
        
        # Add term to the expression
        expr = expr+ a_n * sin(2 * pi * n * x)
    f_numpy=numpify(expr)
    @NamedFunction
    def F(arg):
        return None
    F.f_numpy=f_numpy
    return F
```

```python
NUMBER_OF_MODES = 5

F = make_random_sin_function(modes=NUMBER_OF_MODES)

display(
    LaTeX(
        f"$F(x)$ is a function that has the form $\\sum_{{n=0}}^{NUMBER_OF_MODES} a_n \\sin(2\\pi n x)$ \n with random numbers $a_n\\in [-1,1]$"
    )
)
```

```python
fig = SmartFigure(x_range=(-1 / 2, 1 / 2))
```

```python
fig.plot(F(x), symbol=x, name="Hidden")
```

```python
a=[]
for n in range(5+1):
   a.append(Symbol(f"a_{n}")) 

MSG="Created symbols: "
for an in a:
    MSG=MSG+f"${latex(an)}$ "
display(LaTeX(MSG))
```

```python
for N in [n+1 for n in range(5)]:
    expr=0
    for n in [n+1 for n in range(N)]:
        expr=expr+a[n]*sin(2*pi*n* x)
    MSG=f"Plotting $S_N(x)={latex(expr)}$"
    display(LaTeX(MSG))
    fig.plot( expr, symbol=x, name=f"S_{N}")
```
