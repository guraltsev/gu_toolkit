# Toolkit animation demo

This notebook shows the new per-parameter animation controls. Each slider now has a small play/pause button, and the cogwheel popup includes:

- animation time (seconds per full range, default `5.0`)
- animation mode: `>>` loop, `>` one-shot, `<>` bounce


```python
import sympy as sp
from IPython.display import display

from gu_toolkit import Figure

x, a, b = sp.symbols("x a b")

fig = Figure(title="Animated parameters")
fig.parameter(a, value=0.0, min=0.0, max=2.0, step=0.05)
fig.parameter(b, value=0.0, min=-1.0, max=1.0, step=0.05)
fig.plot(a * sp.sin(x) + b, x, id="animated", label=r"$a\sin(x)+b$")

ref_a = fig.parameters[a]
ref_b = fig.parameters[b]

ref_a.animation_time = 3.0
ref_a.animation_mode = ">>"
ref_b.animation_time = 4.0
ref_b.animation_mode = "<>"

display(fig)
```

Use the small play button beside each parameter to start or pause animation. The controller keeps an internal continuous position, but the displayed value is always snapped to the slider's current `min` / `max` / `step` configuration.


```python
# Programmatic control is available through the ParamRef as well.
ref_a.start_animation()
# ref_a.stop_animation()
# ref_b.toggle_animation()
```
